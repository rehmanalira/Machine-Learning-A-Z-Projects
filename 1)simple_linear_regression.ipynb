# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset=pd.read_csv('ML/Salary_Data.csv')
# using iloc Read From Note Book
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1]
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
#using fit method in to fit our train data
regressor.fit(X_train,y_train)
```

## Predicting the Test set results

```python
# here we use the class  of preditct to of X_test first we test X_test and then y_test
y_pred = regressor.predict(X_test)
```

## Visualising the Training set results

```python
#here we scater those points which are according to the csv file
plt.scatter(X_train, y_train, color='red')

#here we plotig our predicted line which will touches the actual predicted values
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary Vs Experice On TRAIN SET')
plt.xlabel('Number Of years')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```python
#here we scater those points which are according to the csv file
plt.scatter(X_test, y_test, color='red')

#here we plotig our predicted line which will touches the actual predicted values
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary Vs Experice On TEST SET')
plt.xlabel('Number Of years')
plt.ylabel('Salary')
plt.show()
```
